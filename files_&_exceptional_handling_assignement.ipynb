2.  Write a Python program to read the contents of a file and print each line.

```python
def read_and_print_file(filename):
    """
    Reads the contents of a file and prints each line.

    Args:
        filename (str): The path to the file you want to read.
    """
    try:
        # Open the file in read mode ('r')
        with open(filename, 'r') as file:
            print(f"--- Contents of '{filename}' ---")
            # Iterate over each line in the file object
            for line in file:
                # The line variable includes the newline character,
                # so print() will add an extra blank line.
                # Using end='' prevents this, or use line.strip()
                print(line.rstrip())
        print("--- End of file ---")
    except FileNotFoundError:
        print(f"Error: The file '{filename}' was not found.")
    except Exception as e:
        print(f"An error occurred: {e}")

# --- Example Usage ---

# 1. Create a dummy file named 'example.txt' for testing
try:
    with open('example.txt', 'w') as f:
        f.write("This is line 1.\n")
        f.write("This is line 2.\n")
        f.write("This is the final line, line 3.\n")
    print("Created 'example.txt' for demonstration purposes.")
except IOError as e:
    print(f"Could not create dummy file: {e}")

# 2. Call the function with the filename
read_and_print_file('example.txt')

# 3. Example of how to handle a non-existent file
read_and_print_file('non_existent_file.txt')
```

4. Write a Python script that reads from one file and writes its content to another file.

```python
def copy_file_content(source_file, destination_file):
    """
    Reads content from source_file and writes it to destination_file.
    """
    try:
        # Open source file in read mode ('r')
        with open(source_file, 'r') as f_source:
            content = f_source.read()

        # Open destination file in write mode ('w')
        # This will create the file if it doesn't exist or overwrite existing content
        with open(destination_file, 'w') as f_destination:
            f_destination.write(content)

        print(f"Content successfully copied from '{source_file}' to '{destination_file}'.")

    except FileNotFoundError:
        print(f"Error: The source file '{source_file}' was not found.")
    except IOError as e:
        print(f"Error reading or writing file: {e}")

# --- Example Usage ---
# 1. Create a dummy source file for testing
try:
    with open('source.txt', 'w') as f_dummy:
        f_dummy.write("Hello, this is a test line.\nThis is the second line.")
except IOError as e:
    print(f"Could not create dummy source file: {e}")

# 2. Run the copy function
copy_file_content('source.txt', 'destination.txt')

# 3. Optional: Verify the content of the destination file
try:
    with open('destination.txt', 'r') as f_dest_read:
        print("\nContent of destination.txt:")
        print(f_dest_read.read())
except IOError as e:
    print(f"Could not read destination file for verification: {e}")
```

```python
def safe_division(numerator, denominator):
    try:
        result = numerator / denominator
        return result
    except ZeroDivisionError:
        # Handle the error specifically
        print("Error: Cannot divide by zero.")
        return None  # Return a default value or handle as appropriate
    except TypeError:
        # Optionally, handle other potential errors like non-numeric inputs
        print("Error: Ensure both inputs are numbers.")
        return None
    except Exception as e:
        # A general handler for any other unexpected errors
        print(f"An unexpected error occurred: {e}")
        return None

# Example usage:
num = 10
den1 = 2
den2 = 0

print(f"{num} divided by {den1} is: {safe_division(num, den1)}")
print(f"{num} divided by {den2} is: {safe_division(num, den2)}")
```

8. Write a program to handle a file opening error using exception handling.

```python
def read_file_with_exception_handling(filename):
    """
    Attempts to open and read a file, handling common errors gracefully.

    Args:
        filename (str): The path to the file to be read.
    """
    try:
        # Use 'with open()' to ensure the file is automatically closed
        with open(filename, 'r') as file:
            content = file.read()
            print("File read successfully. Content:")
            print(content)
    except FileNotFoundError:
        # Handle the specific error if the file does not exist
        print(f"Error: The file '{filename}' was not found. Please check the file path.")
    except PermissionError:
        # Handle the specific error if the user lacks the necessary permissions
        print(f"Error: Permission denied. You do not have access to the file '{filename}'.")
    except Exception as e:
        # Handle any other unexpected exceptions
        print(f"An unexpected error occurred: {e}")
    else:
        # This block executes only if no exception was raised in the try block
        print("\nNo exceptions occurred during file reading.")
    finally:
        # This block always executes, useful for any final cleanup actions
        print("Finished the file operation attempt.")

# --- Example Usage ---

# 1. Example with a non-existent file
print("--- Attempting to read 'non_existent_file.txt' ---")
read_file_with_exception_handling('non_existent_file.txt')

print("-" * 40)

# 2. Example with a file that actually exists (assuming you have a 'sample.txt' in the same directory)
# For this example to work, create a 'sample.txt' file with some text
print("--- Attempting to read 'sample.txt' ---")
read_file_with_exception_handling('sample.txt')
```

11.  Write a Python program that uses a try-except block to handle an error when attempting to access a
dictionary key that doesn't exist.

```python
def get_student_grade(student_name, grade_book):
    """
    Attempts to retrieve a student's grade from a dictionary using try-except.
    """
    try:
        # Attempt to access the dictionary key
        grade = grade_book[student_name]
        print(f"{student_name}'s grade is: {grade}")
    except KeyError:
        # Handle the specific error if the key is not found
        print(f"Error: '{student_name}' not found in the grade book.")
    except Exception as e:
        # Handle any other unexpected errors
        print(f"An unexpected error occurred: {e}")

# --- Example Usage ---

# A sample dictionary
grades = {
    "Alice": "A",
    "Bob": "B",
    "Charlie": "C"
}

# Accessing an existing key
get_student_grade("Bob", grades)

# Accessing a non-existent key, which triggers the KeyError handler
get_student_grade("David", grades)
```

14. Write a program that uses the logging module to log both informational and error messages.

```python
import logging
import sys

# 1. Configure the logging system
# This setup directs logs to both a file ('app.log') and the console (stdout).
# The 'level=logging.INFO' means that only messages of INFO severity and higher
# (INFO, WARNING, ERROR, CRITICAL) will be processed by the root logger.
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(levelname)s - %(message)s',
    handlers=[
        logging.FileHandler("app.log", mode='w'), # Log to a file, 'w' mode overwrites each run
        logging.StreamHandler(sys.stdout)        # Log to the console (standard output)
    ]
)

# 2. Get a logger instance (using __name__ is a best practice)
logger = logging.getLogger(__name__)

# 3. Log messages at different levels
def divide_numbers(a, b):
    """Attempts to divide two numbers and logs the process and any errors."""
    logger.info(f"Attempting to divide {a} by {b}") # Informational message
    try:
        result = a / b
        logger.info(f"Division successful. Result: {result}")
        return result
    except ZeroDivisionError:
        # Log an error message with exception details
        logger.error("Failed to divide by zero.", exc_info=True)
        return None
    except Exception as e:
        # A generic exception handler
        logger.exception(f"An unexpected error occurred: {e}") # Same as logger.error(exc_info=True)
        return None

# Example usage
logger.info("Program started.")
divide_numbers(10, 2)
divide_numbers(10, 0) # This will trigger the error log
logger.info("Program finished.")
```

17. Write a Python program to create and write a list of numbers to a file, one number per line.

```python
def write_numbers_to_file(filename, numbers):
    """
    Writes a list of numbers to a file, with one number per line.

    Args:
        filename (str): The name of the file to write to.
        numbers (list): A list of numbers (integers or floats).
    """
    try:
        # Open the file in write mode ('w').
        # Using 'with open(...)' ensures the file is automatically closed.
        with open(filename, 'w') as file:
            for number in numbers:
                # Convert the number to a string and add a newline character
                file.write(str(number) + '\n')
        print(f"Successfully wrote numbers to {filename}")
    except IOError as e:
        print(f"An I/O error occurred: {e}")

# Example usage:
if __name__ == "__main__":
    my_list_of_numbers = [10, 20, 30.5, 45, 100]
    file_to_create = "output_numbers.txt"

    write_numbers_to_file(file_to_create, my_list_of_numbers)

    # Optional: Read the file back to verify its contents
    print(f"\nContents of {file_to_create}:")
    try:
        with open(file_to_create, 'r') as file:
            print(file.read())
    except IOError as e:
        print(f"Could not read back the file: {e}")

```

18.  How would you implement a basic logging setup that logs to a file with rotation after 1MB?

```python
import logging
from logging.handlers import RotatingFileHandler
import os

# Define the log file name and maximum size (1MB = 1048576 bytes)
LOG_FILENAME = 'app.log'
MAX_LOG_SIZE = 1048576
BACKUP_COUNT = 5  # Number of backup files to keep (app.log.1, app.log.2, etc.)

# 1. Create a logger instance
logger = logging.getLogger(__name__)
logger.setLevel(logging.DEBUG)  # Set the minimum log level to capture (DEBUG, INFO, WARNING, ERROR, CRITICAL)

# 2. Create the RotatingFileHandler
handler = RotatingFileHandler(
    LOG_FILENAME,
    mode='a',  # Append mode (default)
    maxBytes=MAX_LOG_SIZE,
    backupCount=BACKUP_COUNT,
    encoding='utf-8'
)
handler.setLevel(logging.DEBUG) # Set the handler's minimum log level

# 3. Define the log message format
formatter = logging.Formatter('%(asctime)s - %(name)s - %(levelname)s - %(message)s')

# 4. Add the formatter to the handler
handler.setFormatter(formatter)

# 5. Add the handler to the logger
logger.addHandler(handler)

# --- Example Usage ---
if __name__ == '__main__':
    logger.info("Starting the application")
    for i in range(20):
        # Log messages to test rotation
        logger.debug(f"Logging message number {i}")
        # To see the rotation in action quickly, you could log larger messages
        # or use a much smaller maxBytes value for testing.

    logger.info("Application finished")
```

19. Write a program that handles both IndexError and KeyError using a try-except block.

```python
def handle_errors(data_structure, key_or_index):
    try:
        # Attempt to access an element using either a key (for dictionaries)
        # or an index (for lists/tuples)
        value = data_structure[key_or_index]
        print(f"Successfully accessed value: {value}")

    except (IndexError, KeyError) as e:
        # This block catches both IndexError and KeyError
        print(f"An error occurred: {e}")
        if isinstance(e, IndexError):
            print("Specific error type: IndexError - the index is out of range.")
        elif isinstance(e, KeyError):
            print("Specific error type: KeyError - the key was not found.")
        print("Please check if your key/index exists in the data structure.")

    except Exception as e:
        # Catch any other unexpected exceptions
        print(f"An unexpected error occurred: {e}")

# --- Example Usage ---

# 1. Dictionary example (KeyError expected)
my_dict = {"a": 1, "b": 2}
print("--- Testing KeyError ---")
handle_errors(my_dict, "c") # "c" key does not exist

print("\n" + "="*20 + "\n")

# 2. List example (IndexError expected)
my_list = [10, 20, 30]
print("--- Testing IndexError ---")
handle_errors(my_list, 5) # Index 5 is out of bounds

print("\n" + "="*20 + "\n")

# 3. Successful access example
print("--- Testing Successful Access ---")
handle_errors(my_dict, "a")
```

20.  How would you open a file and read its contents using a context manager in Python?

```python
def read_file_contents(filename):
    """Reads the entire content of a file using a context manager."""
    try:
        # 'r' mode is for reading (default), 'utf-8' is a common encoding
        with open(filename, 'r', encoding='utf-8') as file:
            content = file.read()  # Read all contents into a single string
            print(content)
    except FileNotFoundError:
        print(f"Error: The file '{filename}' was not found.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Example usage:
# Assuming 'example.txt' exists in the same directory
read_file_contents('example.txt')
```

21. Write a Python program that reads a file and prints the number of occurrences of a specific word.

```python
import string
import re

def count_word_occurrences(filename, target_word):
    """
    Reads a file and counts the occurrences of a specific word (case-insensitive, ignores punctuation).
    """
    count = 0
    # Use a context manager to ensure the file is closed automatically
    try:
        with open(filename, 'r', encoding='utf-8') as file:
            for line in file:
                # Convert the line to lowercase to ensure case-insensitive counting
                line = line.lower()
                # Remove punctuation from the line to count words like "hello" and "hello!" as the same
                line = line.translate(str.maketrans('', '', string.punctuation))
                # Split the line into words
                words = line.split()
                # Count the target word in the list of words
                for word in words:
                    if word == target_word.lower():
                        count += 1
    except FileNotFoundError:
        print(f"Error: The file '{filename}' was not found.")
        return -1
    except Exception as e:
        print(f"An error occurred: {e}")
        return -1

    return count

# --- Example Usage ---
if __name__ == "__main__":
    # Create a sample file for demonstration
    with open("sample_file.txt", "w") as f:
        f.write("Hello world, hello there! This is a test file.\n")
        f.write("Hello again, world. This is another line.\n")

    file_name = "sample_file.txt"
    word_to_count = "hello" # The search is case-insensitive

    occurrences = count_word_occurrences(file_name, word_to_count)

    if occurrences != -1:
        print(f"The word '{word_to_count}' appears {occurrences} times in '{file_name}'.")

```

23. Write a Python program that writes to a log file when an error occurs during file handling.

```python
import logging
import os

def setup_logging(log_file='file_errors.log'):
    """
    Configures the logging module to write ERROR messages and higher
    to a specified log file.
    """
    logging.basicConfig(
        filename=log_file,
        level=logging.ERROR,
        format='%(asctime)s - %(levelname)s - %(message)s'
    )

def handle_file_operation(filename, mode, content=None):
    """
    Attempts a file operation (read/write) and logs any errors that occur.
    """
    setup_logging()

    # Use 'with open()' for automatic resource management (cleanup)
    try:
        with open(filename, mode) as f:
            if mode == 'r':
                print(f"Reading from {filename}...")
                data = f.read()
                print("Content read successfully.")
                return data
            elif mode == 'w':
                print(f"Writing to {filename}...")
                if content:
                    f.write(content)
                    print("Content written successfully.")
            else:
                logging.error(f"Unsupported file mode: {mode}")

    except FileNotFoundError:
        # Log a specific error message for a missing file
        logging.error(f"Error: The file '{filename}' was not found.")
        print(f"Error: The file '{filename}' was not found. Check the log file for details.")
    except PermissionError:
        # Log error for permission issues
        logging.exception(f"Error: You do not have permission to access the file '{filename}'.")
        print(f"Error: Permission denied for file '{filename}'. Check the log file for details.")
    except Exception as e:
        # Log any other unexpected exceptions with the full traceback
        logging.exception(f"An unexpected error occurred during file handling: {e}")
        print(f"An unexpected error occurred. Check the log file for details.")
    finally:
        # This block is useful for actions that must run regardless of errors
        print("Attempted file operation cleanup complete.")

# --- Example Usage ---

# 1. Example of a successful file write operation
print("--- Attempting to write a new file ---")
handle_file_operation('test_file.txt', 'w', 'Hello, world! This is a test log entry.')

# 2. Example of an error (trying to read a non-existent file)
print("\n--- Attempting to read a non-existent file ---")
# This will trigger the FileNotFoundError exception and log it
handle_file_operation('non_existent_file.txt', 'r')

# 3. Example of an error (simulating a permission error on a system file path)
# Note: This example assumes a Unix-like system where /root might be protected.
print("\n--- Attempting to access a protected file path ---")
handle_file_operation('/root/protected_file.txt', 'r')

# Clean up the test file if it exists
if os.path.exists('test_file.txt'):
    os.remove('test_file.txt')
```
